```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoding = LabelEncoder()
data['Time'] = encoding.fit_transform(data['Time'])
data['Date'] = encoding.fit_transform(data['Date'])
data['Day of the week'] = encoding.fit_transform(data['Day of the week'])
```

```python
data1 = data.drop(['Traffic Situation'],axis = 1)
data1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data1,data2, test_size = 0.2, random_state = 42)
X_train,X_test,y_train,y_test
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model1 = SVC(kernel='linear')
model1.fit(X_train,y_train)
model1.score(X_test,y_test)
```

```python
model2 = SVC(kernel='poly')
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
```

```python
model3 = SVC(kernel='rbf')
model3.fit(X_train,y_train)
model3.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model4 = LogisticRegression()
model4.fit(X_train,y_train)
model4.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model5 = DecisionTreeClassifier()
model5.fit(X_train,y_train)
model5.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model6 = RandomForestClassifier()
model6.fit(X_train,y_train)
model6.score(X_test,y_test)
```
